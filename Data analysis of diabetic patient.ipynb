```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
from sklearn.linear_model import LogisticRegression
```

```python

mydata=pd.read_csv("https://raw.githubusercontent.com/Premalatha-success/HHE_June5/main/pima-indians-diabetes.csv")
mydata.head()
```

```python
mydata.info()
```

```python
mydata.isnull().sum()
```

```python
mydata["Plas"]=mydata["Plas"].replace(to_replace=0,value=np.nan) # to replace 0 with nan as object datatype
mydata["Plas"]=mydata["Plas"].astype(float)
```

```python
mydata.head(25)
```

```python
mydata.info()
```

```python
mydata["Pres"]=mydata["Pres"].replace(to_replace=0,value=np.nan) #to replace the 0 & convert into object type data
mydata["Pres"]=mydata["Pres"].astype(float)
mydata["skin"]=mydata["skin"].replace(to_replace=0,value=np.nan) #to replace the 0 & convert into object type data
mydata["skin"]=mydata["skin"].astype(float)
mydata["mass"]=mydata["mass"].replace(to_replace=0,value=np.nan) #to replace the 0 & convert into object type data
mydata["mass"]=mydata["mass"].astype(float)
```

```python
mydata.info()
```

```python
median1=mydata["Plas"].median() #removing null values

mydata["Plas"].replace(np.NaN,median1,inplace=True)
median2=mydata["Pres"].median() #removing null values

mydata["Pres"].replace(np.NaN,median2,inplace=True)
median3=mydata["skin"].median() #removing null values

mydata["skin"].replace(np.NaN,median3,inplace=True)
median3=mydata["mass"].median() #removing null values

mydata["mass"].replace(np.NaN,median3,inplace=True)
```

```python
mydata.info()
```

```python
mydata.head(25)
```

```python
from sklearn.model_selection import train_test_split
pd.set_option('display.max_rows',None)
pd.set_option('display.max_columns',None)
```

```python
mydata.info()
```

```python
X=mydata.drop(["Plas"],axis=1)
y=mydata["Plas"]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.3,random_state=1)
```

```python
reg_model=LogisticRegression()
```

```python
reg_model.fit(X_train,y_train)
```

```python
reg_model.score(X_train,y_train) 
```

```python
reg_model.score(X_test,y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model_1=DecisionTreeClassifier()
model_1.fit(X_train,y_train)
```

```python
model_1.score(X_train,y_train)
```

```python
model_1.score(X_test,y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model=RandomForestRegressor()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_train,y_train)
```

```python
model.score(X_test,y_test)
```
